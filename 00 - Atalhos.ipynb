### Há dois modos em uma célula do Jupyter:

1. Edit Mode
2. Command Mode

### Edit Mode

### Command Mode

* Alterando o **Mode**
    * **Esc** $\rightarrow$ **<span style = 'color: blue'>Edit Mode</span>**
    * **Enter** $\rightarrow$ **<span style = 'color: green'>Command Mode</span>**   

* Alterando o tipo de célula do **Edit Mode**
    * **m** $\rightarrow$ Markdown (Texto, imagens e outros tipos de edição)
    * **y** $\rightarrow$ Código

<div class = 'alert alert-block alert-info'>
     <b>Informação</b>: Se você estiver no Command Mode, você conseguirá acessar as células de cima e de baixo atráves das teclas do seu teclado: Page Up e Page Down respectivamente.
</div>

* Inserir ou deletar células (**<span style = 'color: blue'>Edit Mode</span>**)
    * a $\rightarrow$ Adiciona uma célula acima
    * b $\rightarrow$ Adiciona uma célula abaixo
    * d d (Clicando duas vezes rapidamente) $\rightarrow$ Deleta a célula atual

* Execução das células (**<span style = 'color: blue'>Edit Mode</span>**)
    * **Ctrl** + **Enter** $\rightarrow$ Executa a célula atual
    * **Ctrl** + **Shift** + **Enter** $\rightarrow$ Executa todas as células

* Outros atalhos (**Primeiro deve ser selecionado qual ou quais linhas a serem comentadas**) (**<span style = 'color: green'>Command Mode</span>**)
    * **Ctrl** + **/** $\rightarrow$ Comenta a linha (A linha não é executada)
    * **Tab** $\rightarrow$ Identa a linha (Dá um certo espaço em relação a extremidade esquerda) 
    * **Shift** + **Tab** $\rightarrow$ Remove a identação da linha (Recua o espaço de uma identação)

Você pode ir em:

* Help
    * Edit Keyboard Shortcuts

E adicionar outros atalhos.

**Exemplo**

* Vá em **run all cells above** clique em **add shortcut** e escreva **ctrl-shift-up**, clique no **+** e depois em **OK**
* Vá em **run all cells below** clique em **add shortcut** e escreva **ctrl-shift-down**, clique no **+** e depois em **OK**
